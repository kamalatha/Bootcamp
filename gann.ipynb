```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, LeakyReLU, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from google.colab import files
import io
import os
import matplotlib.pyplot as plt
import tensorflow as tf

# 1. Upload your dataset
print("Please upload your dataset in CSV format.")
file_name = 'vibration.csv'
if os.path.exists(file_name):
  os.remove(file_name)
uploaded = files.upload()
if not uploaded:
    raise ValueError("No file was uploaded.")
file_name = next(iter(uploaded))

# 2. Load your dataset
data = pd.read_csv(io.BytesIO(uploaded[file_name]))
print(f"\nSuccessfully loaded '{file_name}'.")

# 3. Preprocessing
data = pd.get_dummies(data, columns=['Surface_Type', 'Material'], drop_first=True)
scaler = MinMaxScaler(feature_range=(-1, 1))
data_scaled = scaler.fit_transform(data)
original_columns = data.columns

# 4. Define GAN Hyperparameters
latent_dim = 100
data_dim = data_scaled.shape[1]
# --- KEY CHANGE: Define separate learning rates ---
generator_lr = 0.0002
discriminator_lr = 0.0001 # Slower learning rate for the discriminator
beta_1 = 0.5
epochs = 10000
batch_size = 128

# 5. Build the Generator Model (Unchanged)
def build_generator():
    model = Sequential(name="Generator")
    model.add(Input(shape=(latent_dim,)))
    model.add(Dense(256))
    model.add(BatchNormalization())
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dense(512))
    model.add(BatchNormalization())
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dense(1024))
    model.add(BatchNormalization())
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dense(data_dim, activation='tanh'))
    return model

# 6. Build the Discriminator Model (Corrected for Stability)
def build_discriminator():
    model = Sequential(name="Discriminator")
    model.add(Input(shape=(data_dim,)))
    # --- KEY CHANGE: Simplified architecture and added Dropout ---
    model.add(Dense(512)) # Simplified from 1024
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dropout(0.4)) # Added Dropout
    model.add(Dense(256)) # Simplified from 512
    model.add(BatchNormalization())
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dropout(0.4)) # Added Dropout
    # The final layers are kept simple
    model.add(Dense(128))
    model.add(LeakyReLU(negative_slope=0.2))
    model.add(Dense(1, activation='sigmoid'))
    return model


# 7. Build Models and Define Optimizers (Corrected)
# --- KEY CHANGE: Use two separate optimizers with different learning rates ---
generator_optimizer = Adam(learning_rate=generator_lr, beta_1=beta_1)
discriminator_optimizer = Adam(learning_rate=discriminator_lr, beta_1=beta_1)

generator = build_generator()
discriminator = build_discriminator()
cross_entropy = tf.keras.losses.BinaryCrossentropy()

print("\n--- Generator Architecture (Unchanged) ---")
generator.summary()
print("\n--- Discriminator Architecture (Improved for Stability) ---")
discriminator.summary()


# 8. Train the GAN using tf.GradientTape
print("\nStarting GAN training with stability improvements...")
d_losses, g_losses = [], []
d_accuracies = []

dataset = tf.data.Dataset.from_tensor_slices(data_scaled).shuffle(buffer_size=data_scaled.shape[0]).batch(batch_size, drop_remainder=True)

@tf.function
def train_step(real_data):
    noise = tf.random.normal([batch_size, latent_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_data = generator(noise, training=True)
        real_output = discriminator(real_data, training=True)
        fake_output = discriminator(generated_data, training=True)

        gen_loss = cross_entropy(tf.ones_like(fake_output), fake_output)

        real_loss = cross_entropy(tf.ones_like(real_output) * 0.9, real_output) # Using label smoothing
        fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
        disc_loss = real_loss + fake_loss

    gen_gradients = gen_tape.gradient(gen_loss, generator.trainable_variables)
    disc_gradients = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gen_gradients, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(disc_gradients, discriminator.trainable_variables))

    # Calculate accuracy
    pred = tf.concat([tf.ones_like(real_output), tf.zeros_like(fake_output)], axis=0)
    res = tf.concat([real_output, fake_output], axis=0)
    acc = tf.reduce_mean(tf.cast(tf.equal(pred, tf.round(res)), tf.float32))

    return gen_loss, disc_loss, acc

for epoch in range(epochs):
    epoch_d_loss, epoch_g_loss, epoch_d_acc = [], [], []
    for batch in dataset:
        g_loss, d_loss, d_acc = train_step(batch)
        epoch_d_loss.append(d_loss)
        epoch_g_loss.append(g_loss)
        epoch_d_acc.append(d_acc)

    d_losses.append(np.mean(epoch_d_loss))
    g_losses.append(np.mean(epoch_g_loss))
    d_accuracies.append(np.mean(epoch_d_acc))

    if (epoch + 1) % 500 == 0:
        print(f"Epoch {epoch + 1}/{epochs}, [D loss: {d_losses[-1]:.4f}, acc: {d_accuracies[-1]*100:.2f}%] [G loss: {g_losses[-1]:.4f}]")

print("GAN training finished.")


# 9. Generate and Display Synthetic Data
print("\n--- Generating Synthetic Data Samples ---")
noise = np.random.normal(0, 1, (10, latent_dim))
generated_scaled_data = generator.predict(noise, verbose=0)
generated_data = scaler.inverse_transform(generated_scaled_data)
synthetic_df = pd.DataFrame(generated_data, columns=original_columns)
for col in [c for c in original_columns if 'Surface_Type' in c or 'Material' in c]:
    synthetic_df[col] = synthetic_df[col].round()
print(synthetic_df.head(10))


# 10. Plot Training Results
plt.figure(figsize=(12, 6))
plt.plot(d_losses, label='Discriminator Loss')
plt.plot(g_losses, label='Generator Loss')
plt.title('GAN Training Losses over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(d_accuracies, label='Discriminator Accuracy')
plt.title('Discriminator Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.axhline(y=0.5, color='r', linestyle='--', label='Ideal Balance (0.5)')
plt.legend()
plt.grid(True)
plt.show()
```
